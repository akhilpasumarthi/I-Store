```
df=pd.read_csv("/content/location.csv")
print(df.head())
print(df.columns)
lat=df['latitude'].tolist()
lon=df['longitude'].tolist()
lat=lat[:50]
lon=lon[:50]

locations=df[['latitude','longitude']]
locationlist=locations.values.tolist()
print(locationlist[7])
```

```
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```
boulder_coords = [28.7041,77.1025]
my_map = folium.Map(location = boulder_coords, zoom_start = 5)

#Add markers to the map
marker_cluster = MarkerCluster().add_to(my_map)  
for i in range(1,50):
 folium.CircleMarker(radius=20,location=locationlist[i],popup='The Waterfront',color='crimson',fill=True,).add_to(marker_cluster)


#Display the map
my_map
```
